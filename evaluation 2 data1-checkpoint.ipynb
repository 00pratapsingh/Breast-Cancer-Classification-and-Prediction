# Independent Variable and Dependent Variable Select

```python
x = rp.iloc[:,1:31]
y = rp.iloc[:,0]
```

```python
x     #Independent variable
```

```python
y              #dependent variable
```

# Data Mapping

```python
y=y.map({'B':0,'M':1})
```

```python
y
```

# Training Data Set Test Data Set

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3,random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression() 
regressor.fit(x_train,y_train)
```

# Prediction

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
 mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score       # R^2 (cofficient of determination)
```

```python
r2_score(y_test,y_pred)                  # R^2 = 1-MSE/MS(avg)
```

```python
from sklearn.preprocessing import PolynomialFeatures 
poly_reg = PolynomialFeatures(degree=2) 
x_poly = poly_reg.fit_transform(x)                                    # y = c + a1x + a2x^2 + a3x^3.....+ anx 
```

```python
lin_reg2 = LinearRegression() 
lin_reg2.fit(x_poly,y)
```

```python
mean_squared_error(y,lin_reg2.predict(poly_reg.fit_transform(x)))
```

```python
r2_score(y,lin_reg2.predict(poly_reg.fit_transform(x)))
```

```python
r2_score(y,regressor.predict(x))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
pred = logmodel.predict(x_test)
```
